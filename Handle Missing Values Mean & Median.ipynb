***Measure of Central Tendency***

- Missing Value Imputation by Mean, Median

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('C:\\Users\\prasad jadhav\\Downloads\\house-prices-advanced-regression-techniques\\train.csv')
```

```python
print(dataset.shape)
print(dataset.head())
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
null_var = dataset.isnull().sum()/dataset.shape[0]*100
null_var
```

```python
drop_columns = null_var[null_var >20].keys()
drop_columns
```

```python
dataset_drop_columns = dataset.drop(columns = drop_columns)
```

```python
dataset_drop_columns.shape
```

```python
dataset_replace = dataset_drop_columns.select_dtypes(include = ['int64', 'float64'])
dataset_replace.head()
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(dataset_replace.isnull())
plt.show()
```

```python
dataset_replace.isnull().sum()
```

```python
null_var = [var for var in dataset_replace if dataset_replace[var].isnull().sum()>0]
null_var
```

```python
plt.figure(figsize = (10, 10))
sns.set()
for i, var in enumerate(null_var):
    plt.subplot(2, 2, i + 1)
    sns.distplot(dataset_replace[var], bins = 20, kde_kws = {'linewidth': 3, 'color': 'red'})
```

- Mean

```python
dataset_update = dataset_replace.fillna(dataset_replace.mean())
dataset_update.isnull().sum().sum()
```

```python
plt.figure(figsize = (10, 10))
sns.set()
for i, var in enumerate(null_var):
    plt.subplot(2, 2, i + 1)
    sns.distplot(dataset_replace[var], bins = 20, kde_kws = {'linewidth': 3, 'color': 'red'}, label = 'Original')
    sns.distplot(dataset_update[var], bins = 20, kde_kws = {'linewidth': 2, 'color': 'green'}, label = 'Mean')
    plt.legend()
```

- Median

```python
dataset_replace_median = dataset_replace.fillna(dataset_replace.median())
dataset_replace_median.isnull().sum().sum()
```

```python
plt.figure(figsize = (10, 10))
sns.set()
for i, var in enumerate(null_var):
    plt.subplot(2, 2, i + 1)
    sns.distplot(dataset_replace[var], bins = 20, hist = False, kde_kws = {'linewidth': 4, 'color': 'red'}, label = 'Original')
    sns.distplot(dataset_update[var], bins = 20, hist = False, kde_kws = {'linewidth': 3, 'color': 'green'}, label = 'Mean')
    sns.distplot(dataset_replace_median[var], bins = 20, hist = False, kde_kws = {'linewidth': 2, 'color': 'blue'}, label = 'Median')
    plt.legend()
```

***Outlier***

```python
for i, var in enumerate(null_var):
    plt.figure(figsize = (10, 10))
    plt.subplot(3, 1, 1)
    sns.boxplot(dataset_replace[var])
    plt.subplot(3, 1, 2)
    sns.boxplot(dataset_update[var])
    plt.subplot(3, 1, 3)
    sns.boxplot(dataset_replace_median[var])
```

```python
dataset_concat = pd.concat([dataset_replace[null_var], dataset_update[null_var], dataset_replace_median[null_var]], axis = 1)
```
