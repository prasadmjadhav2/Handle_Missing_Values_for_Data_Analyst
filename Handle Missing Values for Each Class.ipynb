```python
null_value = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'] # LotFrontage
drop_columns = dataset.drop(columns = null_value)
drop_columns.shape
```

```python
dataset_update = drop_columns.select_dtypes(include = ['int64', 'float64'])
dataset_update.shape
```

```python
dataset_update.isnull().sum()
```

```python
null_var = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
dataset_update[null_var][dataset_update[null_var].isnull().any(axis = 1)]
```

```python
dataset['LotConfig'].unique()
```

```python
dataset[dataset.loc[:, 'LotConfig']== 'Inside']['LotFrontage'].replace(np.nan, dataset[dataset.loc[:, 'LotConfig']== 'Inside']['LotFrontage'].mean())
```

```python
dataset_copy = dataset.copy()

for var_class in dataset['LotConfig'].unique():
    dataset_copy.update(dataset[dataset.loc[:, 'LotConfig']== var_class]['LotFrontage'].replace(np.nan, dataset[dataset.loc[:, 'LotConfig']== var_class]['LotFrontage'].mean()))
```

```python
dataset_copy.isnull().sum()
```

```python
dataset_copy = dataset.copy()

null_var = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_var = ['LotConfig', 'Exterior2nd', 'GarageType']
for cat_var, null_var in zip(cat_var, null_var):
    for var_class in dataset[cat_var].unique():
        dataset_copy.update(dataset[dataset.loc[:, cat_var]== var_class][null_var].replace(np.nan, dataset[dataset.loc[:, cat_var]== var_class][null_var].mean()))
```

```python
dataset_copy[null_var].isnull().sum()
```

```python
dataset_copy = dataset.copy()

null_var = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_var = ['LotConfig', 'MasVnrType', 'KitchenQual']
for cat_var, null_var in zip(cat_var, null_var):
    for var_class in dataset[cat_var].unique():
        dataset_copy.update(dataset[dataset.loc[:, cat_var]== var_class][null_var].replace(np.nan, dataset[dataset.loc[:, cat_var]== var_class][null_var].mean()))
```

```python
dataset_copy[null_var].isnull().sum()
```

- Data Distribution

_Mean_

```python
plt.figure(figsize = (10, 10))
sns.set()
for i, var in enumerate(null_var):
    plt.subplot(2, 2, i + 1)
    sns.distplot(dataset_update[var], bins = 20, kde_kws = {'linewidth': 3, 'color': 'red'}, label = 'Original')
    sns.distplot(dataset_copy[var], bins = 20, kde_kws = {'linewidth': 2, 'color': 'green'}, label = 'Mean')
    plt.legend()
```

_Median_

```python
dataset_copy_median = dataset_copy.copy()

null_var = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
cat_vars = ['LotConfig','Exterior2nd','KitchenQual']
for cat_var, num_var_miss in zip(cat_vars, null_var):
    for var_class in dataset[cat_var].unique():
        dataset_copy.update(dataset[dataset.loc[:, cat_var]== var_class][null_var].replace(np.nan, dataset[dataset.loc[:, cat_var]== var_class][null_var].median()))
```

```python
dataset_copy_median[null_var].isnull().sum()
```

```python
plt.figure(figsize = (10, 10))
sns.set()
for i, var in enumerate(null_var):
    plt.subplot(2, 2, i + 1)
    sns.distplot(dataset_update[var], bins = 20, kde_kws = {'linewidth': 4, 'color': 'red'}, label = 'Original')
    sns.distplot(dataset_copy[var], bins = 20, kde_kws = {'linewidth': 3, 'color': 'green'}, label = 'Mean')
    sns.distplot(dataset_copy_median[var], bins = 20, kde_kws = {'linewidth': 2, 'color': 'blue'}, label = 'Median')
    plt.legend()
```

_Outlier_

```python
for i, var in enumerate(null_var):
    plt.figure(figsize = (10, 10))
    plt.subplot(3, 1, 1)
    sns.boxplot(dataset_update[var])
    plt.subplot(3, 1, 2)
    sns.boxplot(dataset_copy[var])
    plt.subplot(3, 1, 3)
    sns.boxplot(dataset_copy_median[var])
```
