```python
null_var = dataset.isnull().sum()/dataset.shape[0]*100
null_var
```

```python
drop_columns = null_var[null_var >17].keys()
drop_columns
```

```python
dataset_drop_columns = dataset.drop(columns = drop_columns)
```

```python
dataset_drop_columns.shape
```

```python
plt.figure(figsize = (25, 25))
sns.heatmap(dataset_drop_columns.isnull())
plt.show()
```

```python
dataset_update = dataset_drop_columns.dropna()
```

```python
pd.concat([dataset['MSZoning'].value_counts()/dataset.shape[0]*100, 
           dataset_update['MSZoning'].value_counts()/dataset_update.shape[0]*100], 
           axis = 1, keys = ['MSZoning_Org', 'MSZoning_Clean'])
```

```python
def cat_var_dist(var):
    return pd.concat([dataset[var].value_counts()/dataset.shape[0]*100, 
           dataset_update[var].value_counts()/dataset_update.shape[0]*100], 
           axis = 1, keys = [var +'_Org', var +'_Clean'])
```

```python
cat_var_dist('MSZoning')
```
